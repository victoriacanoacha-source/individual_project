### 2. Calculating supply chain concentration based on the importers

This function calculates the following: 

1. **Supplier Share**: This measures how much of global supply that exporter represents, for example, if China is the exporter and the supplier share is 0.25 that means China supplies 25% of global semiconductors
2. **Share Squared**: This is calculated because it is used to compute HHI (Herfindahl-Hirschman Index) which is then needed to calculate the supply concentration index
3. **Supply Concentration**: This is the HHI concentration index. It measures how concentrated the supply chain is and references the importing country

```python
df.dtypes
```

```python
def calculate_supply_concentration(df):   
    # total imports per importer-year
    
    df = df[df["flow"] == "M"].copy()
    total_imports = df.groupby(["importer","year"])["trade_value"].transform("sum")
    
    # supplier share
    df["supplier_share"] = df["trade_value"] / total_imports
    
    # square the shares
    df["share_squared"] = df["supplier_share"] ** 2
    
    # compute HHI per importer-year
    hhi = (
        df.groupby(["importer","year"])["share_squared"]
        .sum()
        .reset_index()
        .rename(columns={"share_squared": "supply_concentration"})
    )
    
    return df.merge(hhi, on=["importer","year"])

df = calculate_supply_concentration(df)
```

```python
df
```

```python
#To make it easier to identify countries that have high concentration it is best to classify each country based on their HHI to identify whether they are diversified, moderately concentrated or highly concentrated

def classify_concentration(supply_concentration):
    if supply_concentration < 0.15:
        return "Diversified"
    elif supply_concentration < 0.25:
        return "Moderate concentration"
    else:
        return "High concentration"

df["concentration_risk"] = df["supply_concentration"].apply(classify_concentration)
```

```python
def main_supplier_by_country_year(df):

    data = df.copy()

    totals = data.groupby(["importer", "year"])["trade_value"].transform("sum")
    data["supplier_share"] = data["trade_value"] / totals

    idx = data.groupby(["importer", "year"])["supplier_share"].idxmax()

    result = data.loc[idx, [
        "importer",
        "year",
        "exporter",
        "trade_value",
        "supplier_share"
    ]].copy()

    result = result.rename(columns={"exporter": "main_supplier"})

    return result.sort_values(["importer", "year"])

main_supplier_df = main_supplier_by_country_year(df)
main_supplier_df["importer"].value_counts()
```

### Scenario 1: China's exports restricted 

The U.S.–China technological rivalry has become a central axis of global economic and geopolitical competition. While the United States continues to lead in frontier innovation—most notably in advanced semiconductors and artificial intelligence (AI)—China has consolidated strengths in large-scale implementation, manufacturing capacity, and control over critical segments of global supply chains. These advantages are especially visible in clean energy technologies and in the processing and refinement of critical minerals and rare earths. 

So imagine now due to this rivalry Chinese semiconductor exports fall by 40% due to trade restrictions or export controls.  What would happen to the current semiconductor supply chain if China restricted exports?


```python
df
```

```python
#We want to see how importing countries are affected so we only focus on imports

imports = df[(df["flow"] == "M") & (df["year"] == 2025)].copy()

#calculate baseline supply so total semiconductor imports from all suppliers 

baseline_supply = imports.groupby("importer")["trade_value"].sum().reset_index(name="baseline_supply")

#identify imports from China 

china_imports = imports[imports["exporter"] == "CHN"]
imports_china = china_imports.groupby("importer")["trade_value"].sum().reset_index(name="imports_china")

#lets simulate the disruption so if chinese exports fall by 40% this means importing countries only receive 60& of the original supply grom China 

imports_china["after_shock_imports"] = imports_china["imports_china"] * 0.6

# merge baseline supply with China imports
impact = baseline_supply.merge(imports_china, on="importer", how="left")

# fill NaN values (countries that do not import from China)
impact["imports_china"] = impact["imports_china"].fillna(0)
impact["after_shock_imports"] = impact["after_shock_imports"].fillna(0)

# compute new total imports after the shock
impact["new_baseline_supply"] = (
    impact["baseline_supply"] - impact["imports_china"] + impact["after_shock_imports"]
)

# calculate absolute loss
impact["absolute_loss"] = impact["baseline_supply"] - impact["new_baseline_supply"]

#calculate percentage supply loss
impact["loss_percent"] = impact["absolute_loss"] / impact["baseline_supply"]
```

```python
def to_hundred_millions(df, column, new_column=None, decimals=2):
    """
    Convert a numeric column to units of 100 million.

    Parameters
    ----------
    df : pandas DataFrame
    column : str
        Column to convert
    new_column : str, optional
        Name of new column (default adds '_100m')
    decimals : int
        Number of decimal places
    """

    if new_column is None:
        new_column = f"{column}_100m"

    df[new_column] = (df[column] / 1e8).round(decimals)

    return df
```
