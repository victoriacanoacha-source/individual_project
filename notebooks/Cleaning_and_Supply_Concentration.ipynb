### 2. Calculating supply chain concentration based on the importers

This function calculates the following: 

1. **Supplier Share**: This measures how much of global supply that exporter represents, for example, if China is the exporter and the supplier share is 0.25 that means China supplies 25% of global semiconductors
2. **Share Squared**: This is calculated because it is used to compute HHI (Herfindahl-Hirschman Index) which is then needed to calculate the supply concentration index
3. **Supply Concentration**: This is the HHI concentration index. It measures how concentrated the supply chain is and references the importing country

```python
df.dtypes
```

```python
def calculate_supply_concentration(df):   
    # total imports per importer-year
    
    df = df[df["flow"] == "M"].copy()
    total_imports = df.groupby(["importer","year"])["trade_value"].transform("sum")
    
    # supplier share
    df["supplier_share"] = df["trade_value"] / total_imports
    
    # square the shares
    df["share_squared"] = df["supplier_share"] ** 2
    
    # compute HHI per importer-year
    hhi = (
        df.groupby(["importer","year"])["share_squared"]
        .sum()
        .reset_index()
        .rename(columns={"share_squared": "supply_concentration"})
    )
    
    return df.merge(hhi, on=["importer","year"])

df = calculate_supply_concentration(df)
```

```python
df
```

```python
#To make it easier to identify countries that have high concentration it is best to classify each country based on their HHI to identify whether they are diversified, moderately concentrated or highly concentrated

def classify_concentration(supply_concentration):
    if supply_concentration < 0.15:
        return "Diversified"
    elif supply_concentration < 0.25:
        return "Moderate concentration"
    else:
        return "High concentration"

df["concentration_risk"] = df["supply_concentration"].apply(classify_concentration)
```

```python
def main_supplier_by_country_year(df):

    data = df.copy()

    totals = data.groupby(["importer", "year"])["trade_value"].transform("sum")
    data["supplier_share"] = data["trade_value"] / totals

    idx = data.groupby(["importer", "year"])["supplier_share"].idxmax()

    result = data.loc[idx, [
        "importer",
        "year",
        "exporter",
        "trade_value",
        "supplier_share"
    ]].copy()

    result = result.rename(columns={"exporter": "main_supplier"})

    return result.sort_values(["importer", "year"])

main_supplier_df = main_supplier_by_country_year(df)
main_supplier_df["importer"].value_counts()
```

### Scenario 1: China's exports restricted 

The U.S.–China technological rivalry has become a central axis of global economic and geopolitical competition. While the United States continues to lead in frontier innovation—most notably in advanced semiconductors and artificial intelligence (AI)—China has consolidated strengths in large-scale implementation, manufacturing capacity, and control over critical segments of global supply chains. These advantages are especially visible in clean energy technologies and in the processing and refinement of critical minerals and rare earths. 

So imagine now due to this rivalry Chinese semiconductor exports fall by 40% due to trade restrictions or export controls.  What would happen to the current semiconductor supply chain if China restricted exports?


```python
df
```

```python
#We want to see how importing countries are affected so we only focus on imports

imports = df[(df["flow"] == "M") & (df["year"] == 2024)].copy()

#calculate baseline supply so total semiconductor imports from all suppliers 

baseline_supply = imports.groupby("importer")["trade_value"].sum().reset_index(name="baseline_supply")

#identify imports from China 

china_imports = imports[imports["exporter"] == "CHN"]
imports_china = china_imports.groupby("importer")["trade_value"].sum().reset_index(name="imports_china")

#lets simulate the disruption so if chinese exports fall by 40% this means importing countries only receive 60& of the original supply grom China 

imports_china["after_shock_imports"] = imports_china["imports_china"] * 0.6

# merge baseline supply with China imports
impact = baseline_supply.merge(imports_china, on="importer", how="left")

# fill NaN values (countries that do not import from China)
impact["imports_china"] = impact["imports_china"].fillna(0)
impact["after_shock_imports"] = impact["after_shock_imports"].fillna(0)

# compute new total imports after the shock
impact["new_baseline_supply"] = (
    impact["baseline_supply"] - impact["imports_china"] + impact["after_shock_imports"]
)

# calculate absolute loss
impact["absolute_loss"] = impact["baseline_supply"] - impact["new_baseline_supply"]

#calculate percentage supply loss
impact["loss_percent"] = impact["absolute_loss"] / impact["baseline_supply"]
```

```python
def to_hundred_millions(df, column, new_column=None, decimals=2):
    """
    Convert a numeric column to units of 100 million.

    Parameters
    ----------
    df : pandas DataFrame
    column : str
        Column to convert
    new_column : str, optional
        Name of new column (default adds '_100m')
    decimals : int
        Number of decimal places
    """

    if new_column is None:
        new_column = f"{column}_100m"

    df[new_column] = (df[column] / 1e8).round(decimals)

    return df
```

Why semiconductor supply chains still become concentrated? 

The reason concentration increases despite diversification incentives is structural: 
- semiconductor manufacturing has very strong economies of scale and technological barriers so building advanced fabrication plants requires:

1. tens of billions of dollars
2. specialised engineering expertise

## Classifying countries into structural risk categories

In our dataset we already have: 

1. supply_concentration
2. loss_percent

From the scatterplot we could indentify the structural risk of each countries but lets add these categories to our dataset by: 

1. Computing the median of supply_concentration
2. Comparing each country (importer) to those medians
3. Assigning the category 

```python
#Calculate Medians to create our threshold 

hhi_median = impact["supply_concentration"].median()
loss_median = impact["loss_percent"].median()

#Create the vulnerability Category 

def classify_vulnerability(row):
    
    if row["supply_concentration"] <= hhi_median and row["loss_percent"] <= loss_median:
        return "Resilient"
    
    elif row["supply_concentration"] > hhi_median and row["loss_percent"] > loss_median:
        return "Highly vulnerable"
    
    elif row["supply_concentration"] <= hhi_median and row["loss_percent"] > loss_median:
        return "China dependent"
    
    else:
        return "Concentrated but safe"

impact["vulnerability_type"] = impact.apply(classify_vulnerability, axis=1)

impact["vulnerability_type"].value_counts()

plt.figure(figsize=(10,6))

plt.scatter(impact["supply_concentration"], impact["loss_percent"], alpha=0.7)

texts = []

for _, row in impact.iterrows():
    texts.append(
        plt.text(
            row["supply_concentration"],
            row["loss_percent"],
            row["importer"],
            fontsize=9
        )
    )

adjust_text(texts, arrowprops=dict(arrowstyle="-", color="gray", lw=0.5))

# median lines
plt.axvline(hhi_median, linestyle="--", color="gray")
plt.axhline(loss_median, linestyle="--", color="gray")

plt.xlabel("Supply Concentration (HHI)")
plt.ylabel("Supply Loss (%)")
plt.title("Supply Concentration vs Vulnerability to Chinese Export Restrictions")

plt.show()
```

This plot shows almost a straight upward line meaning that a higher HHI correlates with higher supply loss so countries tend to fall into only two of the four quadrants; bottom-left meaning low concentration + low loss and top-right  meaning high concentration high loss. And very few countries appeat in the top-eft quadrant meaning diversified but China dependent and bottom-right meaning concentrated but not China dependent. Why does this happen? Because our shock simulation is China-specific (Chinese semiconductor exports fall by 40%). Thus countries that have high supply concentration often also have high dependence on China because China is one of the largest semiconductor exporters and many countries rely heavily on Chinese inputs so the two variables (HHI and Supply loss) become tightly linked. 

HHI ↑ → China share ↑ → Supply loss ↑

This produces the pattern we see

What this means analytically is that my data suggests that countries with highly concentrated semiconductor supplier bases are typically concentrated in Chinese suppliers. A country would show in the bottom-right quadrant (concentrated but low China exposure) if it relied heavily on other countries such as South Korea or Japan instead of China but my dataset suggests few countries have that structure. 

Conclusion: Semiconductor supply chain concentration is strongly linked to reliance on Chinese exports therefore diversification reduces exposure to China shocks.

## Cross-Sectional Analysis (Hypothesis Testing)

From our scatter plot, it is clear that countries are more vulnerable if their supply is highly concentrated on one supplier in this scenario China. But in order to strengthen our analysis on whether vulnerability and supply chain concentration are correlated we should look at the statistical relationship between supply concentration and loss of supply. 

In order to determine the vulnerability of countries we want to explain how much semiconductor supply a country loses when Chinese exports fall which is why the outcome of our regression model is loss_percent. But the most important part of the model is Beta One which tells us how supply loss changes when concentration increases, in simple terms it asks: if supply concentration increases does supply loss increase as well? If beta one is positive then that means that more concentration leads to more vulnerability. We include Epsilon to represent all the other factors affecting supply loss that my model doesnt include these could be; trade policies, regional supply chains, technological specialisation, in simple terms it just represents random variation the model cannot explain. 

Variables: 

- Dependent variable: loss_percent (Y)
- Independent variable: supply_concentration (X)
- Regression model: loss_percenti​=β0​+β1​×supply_concentrationi​+ϵi​

Hypotheses: 

1. H0: B1 = 0 (There is no relationship between supply concentration and vulnerability to a China export shock)
2. H1: B1 > 0 (Countries with more concentrated semiconductor supply chains experience larger supply disruptions when Chinese exports fall)
3. Significance level: 5% (0.05)


```python
import statsmodels.api as sm
```

```python
x = impact["supply_concentration"].tolist()
y = impact["loss_percent"].tolist()

x = sm.add_constant(x)

result = sm.OLS(y, x).fit()
print(result.summary())
```

Results: 

1. B1 (the effect of supply concentration on supply loss) is 0.3158 which means if supply concentration increases by 1 then supply loss increases by 0.316 so countries with more concentrated suppliers experience larger supply disruptions
2. p-value is 0.000 (p < 0.001) so we reject the null hypothesis meaning the relationship between supply concentration and supply loss is statistically significant
3. R-squared (explanatory power) is 0.518 meaning that about 52% of the variation in supply loss across countries is explained by supply concentration

Conclusion: The regression results indicate that countries with more concentrated semiconductor supplier networks experience significantly larger supply disruptions following a simulated reduction in Chinese exports. However, the model does not distinguish between general supplier concentration and dependence on China specifically. So we showed that supplu structure matter not just China exposure but what if we want to add that dimension?

### Adding China share in the regression test 

Even though it seems logical to assume that the high supply losses countries experience are because they rely mostly on China and this scenario carries out a shock to Chinese exports, we still need to verify this relationship. 
We want to know where the risk comes from: does it come from dependence on China specifically (supplier identity risk) or from lack of diversification (network structure risk)? This extended model asks; are countries vulnerable because they depend on China or because their supplier networks are poorly diversified?

There are two types of risk each a different source of vulnerability: 

1. Structural Supply Chain Risk: captured by supply_concentration and measured using HHI
2. Geopolitical Exposure: measures how dependent a country is on China specifically 

```python
## Calculate China Share 
imports_2024 = df[(df["flow"]=="M") & (df["year"]==2024)].copy()

total_imports = imports_2024.groupby("importer")["trade_value"].sum()

china_imports = (
    imports_2024[imports_2024["exporter"]=="CHN"]
    .groupby("importer")["trade_value"]
    .sum()
)

china_share = (china_imports / total_imports).reset_index()
china_share.columns = ["importer","china_share"]
```

```python
impact = impact.merge(china_share, on="importer", how="left")
impact
```

More concentrated suppliers increase vulnerability

H0: β2 = 0

Greater dependence on China increases vulnerability

```python
X = impact[["supply_concentration","china_share"]]
X = sm.add_constant(X)

Y = impact["loss_percent"]

model = sm.OLS(y,X).fit()

print(model.summary())
```

By adding Chinese shares it has shown that vulnerability has little to do with supply concentration, rather it is directly linked to Chinese supply this is because the model assumes that lost chinese supply cannot be replaced so diversification cannot help because other suppliers never increase exports thats why the regression have a r-squared value of 1 meaning that China share completely determines the loss. In reality when China restricts exports other suppliers would try to fill part of the gap so countries like Korea or Japan would increase exports but countries with diversified suppliers adapt more easily so we assume that countries with lower HHI replace more of the lsot supply. 

Thus, to create a more realistic scenario and assumption we should introduce a substitution capactiy based on diversification for example: 

- replacement rate = (1-supply_concentration)

So more diversified countries recover more of the lost supply

```python
# Initial China supply shock
impact["china_loss"] = impact["china_share"] * 0.40

# Replacement capacity based on diversification
impact["replacement"] = impact["china_loss"] * (1 - impact["supply_concentration"])

# Final loss after substitution
impact["adjusted_loss"] = impact["china_loss"] - impact["replacement"]

X = impact[["supply_concentration", "china_share"]]
X = sm.add_constant(X)

y = impact["adjusted_loss"]

model = sm.OLS(y, X).fit()

print(model.summary())
```
