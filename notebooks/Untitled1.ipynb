```python
import pandas as pd

def format_columns(df: pd.DataFrame, rename_dict: dict = None, keep_columns: list = None) -> pd.DataFrame:
    df = df.copy()
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(r"[^\w\s]", "", regex=True)
        .str.replace(" ", "_"))
    if rename_dict:
        df = df.rename(columns=rename_dict)
    if keep_columns:
        df = df[keep_columns]
    return df

rename_dict = {
    "refyear": "year",
    "reporteriso": "importer",
    "partneriso": "exporter",
    "flowcode": "flow",
    "cmdcode": "hs_code",
    "cmddesc": "product",
    "primaryvalue": "trade_value",
"motdesc": "transport"}

keep_columns = [
    "year",
    "importer",
    "flow",
    "exporter",
    "transport",
    "hs_code",
    "product",
    "trade_value"]

```

```python
df1 = format_columns(df1, rename_dict, keep_columns)
df2 = format_columns(df2, rename_dict, keep_columns)
df3 = format_columns(df3, rename_dict, keep_columns)
df4 = format_columns(df4, rename_dict, keep_columns)
df5 = format_columns(df5, rename_dict, keep_columns)
df6 = format_columns(df6, rename_dict, keep_columns)
```

## Identifying Supply Chain Concentration: 

```python
supplier_share = (
    df.groupby(["exporter"])["trade_value"]
    .sum()
    / df["trade_value"].sum()).sort_values(ascending=False).head(10)

supplier_share.plot(kind="bar")
plt.show()
```

```python
def calculate_supply_concentration(df):
    # total imports per importer
    total_imports = df.groupby("importer")["trade_value"].transform("sum")
    
    # supplier share for each row
    df["supplier_share"] = df["trade_value"] / total_imports
    
    # square the shares
    df["share_squared"] = df["supplier_share"] ** 2
    
    # compute HHI per importer
    hhi = df.groupby("importer")["share_squared"].sum().reset_index()
    hhi = hhi.rename(columns={"share_squared": "supply_concentration"})
    
    return df.merge(hhi, on="importer")

df = calculate_supply_concentration(df)
```

```python
df
```

```python
def classify_concentration(hhi):
    if hhi < 0.15:
        return "Diversified"
    elif hhi < 0.25:
        return "Moderate concentration"
    else:
        return "High concentration"

df["concentration_risk"] = df["supply_concentration"].apply(classify_concentration)
```

```python
country_concentration = df[["importer", "supply_concentration"]].drop_duplicates()
country_concentration["risk_level"] = country_concentration["supply_concentration"].apply(classify_concentration)
```

```python
high = country_concentration[country_concentration["risk_level"] == "High concentration"]
moderate = country_concentration[country_concentration["risk_level"] == "Moderate concentration"]
diversified = country_concentration[country_concentration["risk_level"] == "Diversified"]
```

```python
high = high.sort_values("supply_concentration", ascending=False)

plt.figure(figsize=(10,5))
plt.bar(high["importer"], high["supply_concentration"], color="red")

plt.title("Countries with High Semiconductor Supply Concentration")
plt.ylabel("HHI")
plt.xlabel("Country")

plt.xticks(rotation=90)
plt.show()
```

```python
moderate = moderate.sort_values("supply_concentration", ascending=False)

plt.figure(figsize=(10,5))
plt.bar(moderate["importer"], moderate["supply_concentration"], color="orange")

plt.title("Countries with Moderate Semiconductor Supply Concentration")
plt.ylabel("HHI")
plt.xlabel("Country")

plt.xticks(rotation=90)
plt.show()
```

```python
diversified = diversified.sort_values("supply_concentration", ascending=False)

plt.figure(figsize=(100,5))
plt.bar(diversified["importer"], diversified["supply_concentration"], color="green")

plt.title("Countries with Diversified Semiconductor Supply Chains")
plt.ylabel("HHI")
plt.xlabel("Country")

plt.xticks(rotation=90)
plt.show()
```
