```python
import pandas as pd

def format_columns(df: pd.DataFrame, rename_dict: dict = None, keep_columns: list = None) -> pd.DataFrame:
    df = df.copy()
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(r"[^\w\s]", "", regex=True)
        .str.replace(" ", "_"))
    if rename_dict:
        df = df.rename(columns=rename_dict)
    if keep_columns:
        df = df[keep_columns]
    return df

rename_dict = {
    "refperiodid": "year",
    "reporteriso": "importer",
    "partneriso": "exporter",
    "flowcode": "flow",
    "cmdcode": "hs_code",
    "cmddesc": "product",
    "primaryvalue": "trade_value",
"motdesc": "transport"}

keep_columns = [
    "year",
    "importer",
    "flow",
    "exporter",
    "transport",
    "hs_code",
    "product",
    "trade_value"]

```

```python
df1 = format_columns(df1, rename_dict, keep_columns)
df2 = format_columns(df2, rename_dict, keep_columns)
df3 = format_columns(df3, rename_dict, keep_columns)
df4 = format_columns(df4, rename_dict, keep_columns)
df5 = format_columns(df5, rename_dict, keep_columns)
df6 = format_columns(df6, rename_dict, keep_columns)
```

```python
df6
```

```python
datasets = {
    "Phase1": df1,
    "Phase2": df2,
    "Phase3": df3,
    "Phase4": df4,
    "Phase5": df5,
    "Phase6": df6
}

for name, df in datasets.items():
    print(f"\n===== {name} =====")
    
    for col in df.columns:
        print(f"\nColumn: {col}")
        print(df[col].value_counts().head())
```

```python
df1["product"] = "semiconductor devices"
df2["product"] = "semiconductor devices"
df3["product"] = "semiconductor devices"
df4["product"] = "semiconductor devices"
df5["product"] = "semiconductor devices"
df6["product"] = "semiconductor devices"
    
```

```python
def check_null(df):
    nulls = df.isna().sum()
    percent = (nulls / len(df)) * 100
    return pd.DataFrame({
        "missing_values": nulls,
        "percent_missing": percent})

print(check_null(df1))
print(check_null(df2))
print(check_null(df3))
print(check_null(df4))
print(check_null(df5))
print(check_null(df6))
```
