```python
import pandas as pd

def format_columns(df: pd.DataFrame, rename_dict: dict = None, keep_columns: list = None) -> pd.DataFrame:
    df = df.copy()
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(r"[^\w\s]", "", regex=True)
        .str.replace(" ", "_"))
    if rename_dict:
        df = df.rename(columns=rename_dict)
    if keep_columns:
        df = df[keep_columns]
    return df

rename_dict = {
    "refyear": "year",
    "reporteriso": "importer",
    "partneriso": "exporter",
    "flowcode": "flow",
    "cmdcode": "hs_code",
    "cmddesc": "product",
    "primaryvalue": "trade_value",
"motdesc": "transport"}

keep_columns = [
    "year",
    "importer",
    "flow",
    "exporter",
    "transport",
    "hs_code",
    "product",
    "trade_value"]

```

```python
df1 = format_columns(df1, rename_dict, keep_columns)
df2 = format_columns(df2, rename_dict, keep_columns)
df3 = format_columns(df3, rename_dict, keep_columns)
df4 = format_columns(df4, rename_dict, keep_columns)
df5 = format_columns(df5, rename_dict, keep_columns)
df6 = format_columns(df6, rename_dict, keep_columns)
```

```python
datasets = {
    "Phase1": df1,
    "Phase2": df2,
    "Phase3": df3,
    "Phase4": df4,
    "Phase5": df5,
    "Phase6": df6
}

for name, df in datasets.items():
    print(f"\n===== {name} =====")
    
    for col in df.columns:
        print(f"\nColumn: {col}")
        print(df[col].value_counts().head())
```

```python
def check_null(df):
    nulls = df.isna().sum()
    percent = (nulls / len(df)) * 100
    return pd.DataFrame({
        "missing_values": nulls,
        "percent_missing": percent})

print(check_null(df1))
print(check_null(df2))
print(check_null(df3))
print(check_null(df4))
print(check_null(df5))
print(check_null(df6))
```

## Merging Datasets

```python
df = pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True)
```

```python
df = (df.groupby(["year","importer","exporter","hs_code","flow"], as_index=False).agg({"trade_value":"sum"}))
```

```python
df
```

```python
def assign_phase(year):
    if year <= 2001:
        return "Early Globalisation"
    if year <= 2008:
        return "Hyper-globalisation"
    if year <= 2017:
        return "Post-financial crisis adjustment"
    if year <= 2018:
        return "Geopolitical fragmentation"
    if year <= 2021:
        return "Global supply chain shock"
    else:
        return "Strategic diversification"
        
df["phase"] = df["year"].apply(assign_phase)
df
```

## Exploratory Data Analysis: 

### Trade Volume:
#### By Phase: 

```python
import plotly as plt
import seaborn as sns
import matplotlib.pyplot as plt

tv_phase = df.groupby("phase")["trade_value"].agg("sum")

tv_phase.plot(kind="bar")
plt.show()
```

Insight: As we can see during the post-financial crisis adjustment period (2009-2017) trade was at its highest however, this could be because this period has more years combined

#### By Year: 

```python
tv_year = df.groupby("year")["trade_value"].agg("sum")
tv_year.plot(kind="bar")
plt.show()

sns.pairplot(df[["year","trade_value"]]);
```
